```python
from tensorflow.keras.applications.mobilenet import MobileNet
from keras.optimizers import Adam
from keras.layers import *
from keras.models import Model
```

```python
model = MobileNet(input_shape=(224,224,3),weights="imagenet",include_top=False)
```

```python
model.summary()
model.input
model.output
```

```python
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation="relu")(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(10,activation="softmax")(d1)

new_model = Model(model.input,fc2)
```

```python
new_model.summary()
new_model.output
```

```python
adam = Adam(learning_rate=0.00003)
# new_model.compile(loss="categorical_crossentropy",optimizer=adam,metrics=["accuracy"])
```

```python
for ix in range(len(new_model.layers)):
    print(ix,new_model.layers[ix])
```

```python
for ix in range(80):
    new_model.layers[ix].trainable = False
new_model.compile(loss="categorical_crossentropy",optimizer=adam,metrics=["accuracy"])
```

```python
new_model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale=1.0/255)
train_generator = datagen.flow_from_directory("D:\Aiml\Data Science CB\\67\\train",target_size=(224,224),batch_size=32)
```

```python
hist = new_model.fit(train_generator,epochs=10,steps_per_epoch=45)
```

```python
y_test = []
new_model.output
```

```python
from keras.utils import load_img,img_to_array
import os
import numpy as np
for i in os.listdir("D:\Aiml\Data Science CB\\67\images"):
    img = load_img("D:\Aiml\Data Science CB\\67\images\\"+i,target_size=(224,224))
    img = img_to_array(img)/255.0
    img = np.expand_dims(img,axis=0)
    p = new_model.predict(img)
    y_test.append(np.argmax(p,axis=-1))
```
